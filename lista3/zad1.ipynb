```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.cluster import KMeans, MiniBatchKMeans, Birch, DBSCAN
```

```python
import matplotlib.colors as colors
from itertools import cycle
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn import datasets
import scipy.spatial.distance as dist

def plotClustering(X, plt_labels):
    plt_colors = cycle(colors.cnames.keys())

    plt_K = np.unique(plt_labels).size

    for k in xrange(plt_K):
        color = plt_colors.next()
        mask = (plt_labels == k)
        plt.plot(X[mask, 0], X[mask, 1], 'w', markerfacecolor=color, marker='o')
```

```python
from sklearn import datasets

centers_ = [[1, 1], [3, 3], [5, 1]]
X, labels = datasets.make_blobs(n_samples=3000, n_features=2, centers=centers_, cluster_std=0.5)

plotClustering(X, labels)
plt.show()
```

```python
birch = Birch(threshold=0.25, n_clusters=None)
birch.fit(X)
#print birch.root_.centroids_
plotClustering(X, birch.labels_)
plt.scatter( birch.root_.centroids_[0, :], birch.root_.centroids_[1, :], marker='o')
plt.show()
```

```python
birch1 = Birch(threshold=0.25, n_clusters=None)
birch1.fit(X)
birch2 = Birch(threshold=0.25, n_clusters=3)
birch2.fit(X)

plotClustering(X, birch1.labels_)
plt.show()
plt.figure(2)
plotClustering(X, birch2.labels_)
plt.show()
```
