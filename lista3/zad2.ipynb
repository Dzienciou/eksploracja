```python
% matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.colors as colors
from itertools import cycle
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn import datasets
import scipy.spatial.distance as dist

from sklearn.cluster import KMeans, MiniBatchKMeans, Birch, DBSCAN
```

```python
def normalize(X):
    return (X - X.mean(axis=0, keepdims=True)) / (X.max(axis=0, keepdims=True) - X.min(axis=0, keepdims=True))
def standarize(X):
    m = X.mean(axis=0, keepdims=True)
    return (X- m) / np.std(X,axis=0)
def plotClustering(X, plt_labels):
    plt_colors = cycle(colors.cnames.keys())

    plt_K = np.unique(plt_labels).size

    for k in xrange(plt_K):
        color = plt_colors.next()
        mask = (plt_labels == k)
        plt.plot(X[mask, 0], X[mask, 1], 'w', markerfacecolor=color, marker='o')

    plt.show()
```

```python
data = np.array([[0, -10], [2, -10], [4, -10], [0, 10], [2, 10], [4, 10], [1, 1000]])

plt.scatter(data[:, 0], data[:, 1])
plt.show()
plt.figure(1)

norm_data=normalize(data)
std_data = standarize(data)
print(std_data)
plt.scatter(norm_data[:, 0], norm_data[:, 1])

plt.show()

plt.figure(2)
plt.scatter(std_data[:, 0], std_data[:, 1])

plt.show()
```

```python

k_means = KMeans(init='k-means++', n_clusters=3, n_init=1)
k_means.fit(data)
plotClustering(data, k_means.labels_)
k_means.fit(norm_data)
plotClustering(norm_data, k_means.labels_)
k_means.fit(std_data)
plotClustering(std_data, k_means.labels_)
```
