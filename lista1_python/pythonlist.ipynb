```python
newTarget = np.array(iris.target[iris.target!=1])
newLength = np.array(iris.data[:, 0][iris.target!=1])
newWidth = np.array(iris.data[:, 1][iris.target!=1])

isInCorrectPart0 = np.array(np.logical_and([newWidth>2*newLength-8], [newTarget == 0]))
isInCorrectPart2 = np.array(np.logical_and([newWidth<2*newLength-8], [newTarget == 2]))
isInCorrectPart = np.logical_or(isInCorrectPart0, isInCorrectPart2)
colors = np.where(isInCorrectPart, "green", "red")


plt.scatter(iris.data[:, 0][iris.target!=1], iris.data[:, 1][iris.target!=1], c = colors.flatten())
x = np.linspace(3.0, 9.0, 100)
plt.plot(x, 2*x-8)

plt.xlabel(iris.feature_names[1])
plt.ylabel(iris.feature_names[0])
plt.xlim((3.0, 9.0))
plt.ylim((1.0, 5.0))
plt.savefig("zadanie2a.png")
plt.show()
```

```python
def centers(k, c):
    angle = 2*math.pi/(k*1.0)
    a = 0.5*c/math.sin(0.5*angle)
    rotate = np.array([[math.cos(angle), -math.sin(angle)],[math.sin(angle), math.cos(angle)]])
    startPoint = np.array([0, a]).flatten()
    points = [np.dot(np.linalg.matrix_power(rotate, i), startPoint).flatten() for i in range(k)]
    return np.array(points)
    
fig = plt.figure(figsize=(3,3)) # default is (8,6)
points = centers(5, 1)
print points
plt.scatter(points[:, 0], points[:, 1])
plt.xlim((-1.5, 1.5))
plt.ylim((-1.5, 1.5))
plt.show()
```

```python
def make_groups(avector, bvector):
    aRands = np.random.normal(size=(avector.size, 1000))
    bRands = np.random.normal(size=(bvector.size, 1000))
    
    pointsx = aRands+avector.reshape(-1, 1)
    pointsy = bRands+bvector.reshape(-1, 1)
    return pointsx, pointsy

def make_groups_figure(K, d):
    cent = centers(K, d)
    return make_groups(cent[:, 0], cent[:, 1])

i=1

for K, d in [(7, 5), (11, 10), (23, 15)]:
    pointsx, pointsy = make_groups_figure(K, d)
    plt.scatter(pointsx, pointsy)
    i=i+1
    plt.figure(i)

plt.show()
```
