```python
from scipy import ndimage
from scipy import stats

import numpy as np
import scipy.spatial.distance as dist

import matplotlib
import matplotlib.pyplot as plt

from collections import Counter
```

```python
f = open('data/kosarak.dat', 'r')
x = f.readlines()
data = [[int(s) for s in str.split(' ')] for str in x]
print data[0:100]
```

```python
c = Counter(x for xs in data for x in xs)
```

```python
def letter_cmp(a, b):
    if a[1] > b[1]:
        return -1
    elif a[1] == b[1]:
        if a[0] > b[0]:
            return 1
        else:
            return -1
    else:
        return 1
c_list = c.items()
c_list.sort(letter_cmp)
best = c.most_common(1000)

best_array=np.array([np.array(xs) for xs in best])
print best_array

plt.figure(figsize=(50, 100))
plt.bar(best_array[:, 0], best_array[:, 1], width = 10, color=['red', 'blue'])
plt.show()
print best
```
